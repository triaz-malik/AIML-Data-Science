### Most influential words (model coefficients)
Positive vs. negative drivers — a cheap, honest explainability check.

```python
import joblib, numpy as np
from src import config
model = joblib.load(config.MODELS_DIR / 'logistic.pkl')
vec, clf = model.named_steps['tfidf'], model.named_steps['clf']
names = np.array(vec.get_feature_names_out())
coef = clf.coef_[0]
top_pos = names[np.argsort(coef)[-20:]][::-1]
top_neg = names[np.argsort(coef)[:20]]
print('PUSH POSITIVE:', list(top_pos))
print('PUSH NEGATIVE:', list(top_neg))
```

### Try it on your own text

```python
samples = ['Absolutely love this, works perfectly!',
           'Broke after two days, complete waste of money.']
for s in samples:
    pred = model.predict([s])[0]
    print(config.LABEL_NAMES[pred], '<-', s)
```
