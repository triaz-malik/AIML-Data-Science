### Compare all three models
Reads the metrics JSON each trainer writes and tabulates them.

```python
import json, pandas as pd
from src import config
rows = []
for f, name in [('logistic_metrics.json','tfidf_logreg'),
                ('distilbert/metrics.json','distilbert'),
                ('bert/metrics.json','bert')]:
    p = config.MODELS_DIR / f
    if p.exists():
        rows.append(json.loads(p.read_text()))
pd.DataFrame(rows)[['model','accuracy','f1']].sort_values('accuracy')
```
