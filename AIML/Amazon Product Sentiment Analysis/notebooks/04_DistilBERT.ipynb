# 04 · DistilBERT

Fine-tune `distilbert-base-uncased`. Lighter/faster than BERT, expected accuracy ≈ 91–93%. Needs a CUDA GPU for reasonable speed.

Logic is in `src/train_transformer.py` (also runnable as a script:
`python -m src.train_transformer --model distilbert-base-uncased --sample 100000`).

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
%matplotlib inline
```

```python
import torch; print('CUDA available:', torch.cuda.is_available())
```

```python
from src import train_transformer
metrics = train_transformer.train(
    model_name='distilbert-base-uncased',
    sample_size=100_000, epochs=2, batch_size=32,
    lr=2e-5, max_len=256, weight_decay=0.01,
)
metrics
```
